### Step 2: Load Raw Data

```python
from src.data.loader import DataLoader

loader = DataLoader()
properties, customers = loader.load_both()

print(f"Properties: {properties.shape}")
print(f"Customers: {customers.shape}")
```

### Step 3: Clean Data

```python
from src.data.cleaner import DataCleaner

cleaner = DataCleaner()
properties_clean, customers_clean = cleaner.clean_all(properties, customers)

print("Data cleaning complete!")
```

### Step 4: Merge Datasets

```python
merged_df = pd.merge(properties_clean, customers_clean, on='customerid', how='inner')

print(f"Merged shape: {merged_df.shape}")
```

### Step 5: Remove Unnamed Columns

```python
unnamed_cols = [col for col in merged_df.columns if 'unnamed' in col.lower()]
if unnamed_cols:
    merged_df = merged_df.drop(columns=unnamed_cols)
    print(f"Dropped: {unnamed_cols}")

print(f"Final shape: {merged_df.shape}")
```

### Step 6: Sanity Check

```python
print("="*50)
print("SANITY CHECK")
print("="*50)
print(f"Expected: 267 rows, 19+ columns")
print(f"Actual: {merged_df.shape[0]} rows, {merged_df.shape[1]} columns")

if merged_df.shape[0] >= 190 and merged_df.shape[1] >= 19:
    print("✅ SANITY CHECK PASSED")
else:
    print("⚠️ Review merge results")
```

```python
output_path = Path("data/processed/merged_real_estate.csv")
output_path.parent.mkdir(parents=True, exist_ok=True)
merged_df.to_csv(output_path, index=False)

print(f"✅ Saved to: {output_path}")
```
